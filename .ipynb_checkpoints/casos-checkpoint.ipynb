```julia
# v0 é o coeficiente de Poisson
function caso_a(v0)
    
    nnos = 4
    ne = 1
    
    #quantos graus de liberdade tenho por no:
    glsno = 2

    #variavel auxiliar que determina as dimensões do problema. N total de graus de liberdade
    dim = nnos*glsno 

    coord = [0.0   0.0   ; #n1
             2.0   0.0   ; #n2
             2.0   2.0   ; #n3
             0.0   2.0   ] #n4                      

    
    conectividades = [1 2 3 4]

    VE =    5000*ones(ne) #Modulo de elasticidade do elemento
    Vnuxy = v0*ones(ne) #Poisson do elemento.
    
    E = VE[1]
    ν = Vnuxy[1]

    #Vou precisar para montar a matriz M.
    λ = (ν*E) / ((1+ν)*(1-2*ν))

    # Espessura
    esp = 1.0

    # Apoios (cond. de contorno essenciais)
    #        no gl valor
    apoios = [1 2 0.0;
              2 2 0.0;
              3 2 0.0;
              4 1 0.0; 
              4 2 0.0 ]

    # Forças (cond. de contorno naturais)
    #         no gl valor
    P = 1000
    forcas = [1  1 -P ;
              2  1  P
              3  1 -P ]
    
    #originalmente devolve espessura
    return nnos, ne,dim,coord,conectividades,VE,Vnuxy,apoios,forcas,λ
end
```

```julia
#Estutura com 3 elementos de base diferente da altura.

function caso_f(v0)
    
    nnos = 8
    ne = 3
    
    #quantos graus de liberdade tenho por no:
    glsno = 2
    b = 2.0

    #variavel auxiliar que determina as dimensões do problema. N total de graus de liberdade
    dim = nnos*glsno 
             #x    #y
    coord = [b     0.0    ; #n1
             b     b/3    ; #n2
             b     2*b/3  ; #n3
             b      b     ; #n4
             0.0   0.0    ; #n5
             0.0   b/3    ; #n6  
             0.0  2*b/3   ; #n7
             0.0    b     ] #n8                                

    
    conectividades = [1 2 6 5;
                      2 3 7 6;
                      3 4 8 7]

    VE =    ones(ne) #Modulo de elasticidade do elemento
    Vnuxy = v0*ones(ne) #Poisson do elemento.
    
    E = VE[1]
    ν = Vnuxy[1]

    #Vou precisar para montar a matriz M.
    λ = (ν*E) / ((1+ν)*(1-2*ν))


    # Apoios (cond. de contorno essenciais)
    #        no gl valor
    apoios = [5 1 0.0;
              5 2 0.0;
              1 2 0.0]

    # Forças (cond. de contorno naturais)
    #         no gl valor
    P = 1
    forcas = [4  2  P;
              8  2  P]
    

    return nnos, ne,dim,coord,conectividades,VE,Vnuxy,apoios,forcas,λ
end
```
